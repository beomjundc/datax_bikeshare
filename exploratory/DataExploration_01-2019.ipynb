```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data = pd.read_csv('201901-fordgobike-tripdata.csv')
```

```python
data.head()
```

```python
data['start_time'] = pd.to_datetime(data['start_time'])
data['end_time'] = pd.to_datetime(data['end_time'])
data = data.sort_values(by='start_time',ascending=True)
data.head()
```

```python
data.describe()
```

```python
plt.figure()
fig, ax = plt.subplots()
fig.set_size_inches(12, 8)

sf_data = data[(data['start_station_longitude'] < -122.35) & (data['start_station_latitude'] > 37.65)]
ax.scatter(sf_data['start_station_longitude'],sf_data['start_station_latitude'],\
           s=10*(sf_data['duration_sec']-np.mean(sf_data['duration_sec']))/(np.std(sf_data['duration_sec'])))
ax.set(title = 'Ride Start Positions for San Francisco', xlabel='Longitude', ylabel='Latitude',\
       xlim=(-122.46,-122.38), ylim=(37.72,37.82));
```

```python
plt.figure()
fig, ax = plt.subplots()
fig.set_size_inches(15, 8)

ax.plot(sf_data['start_time'],sf_data['duration_sec'])
ax.set(title = 'San Francisco Ride Times Throughout January, 2019', xlabel='Date', ylabel='Duration (sec)');
```

```python
data['euclidean_distance'] = ((data['end_station_latitude']-data['start_station_latitude'])**2 \
 + (data['end_station_longitude']-data['start_station_longitude'])**2)**(1/2)
data.head()
```

```python
plt.scatter(data['duration_sec'],data['euclidean_distance']);
plt.title('Ride Time vs Distance')
plt.xlabel('Duration (sec)')
plt.ylabel('Distance (Long x Lat)');
```

```python
data[data['euclidean_distance']>0.2]
```

```python
data_short = data[data['duration_sec']<2000]
plt.scatter(data_short['duration_sec'],data_short['euclidean_distance'])
plt.title('Ride Time vs Distance (<2000 Seconds)')
plt.xlabel('Duration (sec)')
plt.ylabel('Distance (Long x Lat)');
```

```python
plt.scatter(2019 - data['member_birth_year'][data['member_birth_year']>1920], \
            data['duration_sec'][data['member_birth_year']>1920])
plt.title('Ride Time vs User Age')
plt.xlabel('User Age')
plt.ylabel('Duration (sec)');
```
